```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import matplotlib.ticker as mtick
import numpy as np
```

```python
#Read in data
data = pd.read_csv("C:/Users/ericd/OneDrive - North Carolina State University/Desktop/PersonalProjects/WBB/NC State '22-'23/shot_quality.csv")
```

```python
#Remove Sophie Hart(transfer)
data = data[data['Players']!='Sophie Hart']
```

```python
#Establish image paths for function to reference when plotting
data['path'] = "C:/Users/ericd/OneDrive - North Carolina State University/Desktop/PersonalProjects/WBB/NC State '22-'23/Headshots/"+ data['Players']  + '.png'
```

```python
#create plot area
#fig, ax = plt.subplots()

#set plot size
#fig.set_size_inches(7,5)
```

```python
#create image function
def getImage(path):
    return OffsetImage(plt.imread(path), zoom=0.1, alpha=1)
```

```python
#plot
fig, ax = plt.subplots(figsize=(6, 4), dpi=600)
ax.scatter(data['Good_Possession_Rate'], data['Rim_and_3_rate'], color='white')
ax.set_title('NCSU WBB Shot Quality', size=12, fontweight='bold')
ax.set_xlabel('% Good Possessions')
ax.set_ylabel('Rim/3Pt Rate')
plt.xlim(.29,.65)
plt.ylim(.43,1.0)
ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0, decimals=0))
ax.xaxis.set_major_formatter(mtick.PercentFormatter(1.0, decimals=0))

#average ratings
plt.axhline(y=.709, color = 'black', linestyle='dashed', alpha=.5)
fig.text(.78,.5, 'Team Rim/3Pt Rate', size=5)

plt.axvline(x=.407, color = 'black', linestyle='dashed', alpha=.5)
fig.text(.35,.12, 'Team Good Possession %', size=5, rotation=90)

#source and author cred
fig.text(.605,.122, 'Source: www.shotquality.com/player-stats', size=6)
fig.text(.13,.85, 'Created by: Eric Drew',size=6)

#add images
for index, row in data.iterrows():
    ab = AnnotationBbox(getImage(row['path']), (row['Good_Possession_Rate'], row['Rim_and_3_rate']), frameon=False)
    ax.add_artist(ab)
    
#save image
plt.savefig("C:/Users/ericd/OneDrive - North Carolina State University/Desktop/PersonalProjects/WBB/NC State '22-'23/qualityPlot.jpeg",
           edgecolor='black', dpi=500, transparent=True)
```
